```python
import json
import string
```

## Load source rationale

```python
sr = "For example, mASPC2 and hASPC2 are characterized by high expression of Aldh1a3 and ALDH1A3, respectively, and strongly resemble previously identified early multipotent progenitor cells that reside in the reticular interstitium of the fat pad5."
folder = "adipose_Emont2022"
```

## Load ctmap

```python
ctmap_fn = f"../../data/{folder}/ctmap/rev_ctmap.json"

with open(ctmap_fn, 'r') as f:
    rev_ctmap = json.load(f)
```

## Translate into machine-readable phrase

```python
# adding noncontiguous extraction tool in the future (keep naive replacement for now)

text_list = sr.upper().split()
translator = str.maketrans('', '', string.punctuation)

for idx in range(len(text_list)):
    text_list[idx] = text_list[idx].translate(translator)
    
for key in rev_ctmap.keys():
    if key in text_list:
        key_idx = text_list.index(key)
        text_list[key_idx] = rev_ctmap[key]
    
translated_sr = " ".join(text_list).lower()
```
